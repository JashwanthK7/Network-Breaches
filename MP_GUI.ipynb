```python
from tkinter import *;
from tkinter import filedialog
from pickle import load
import pandas as pd
import pandastable as pt
from pandastable import Table
from sklearn.preprocessing import LabelEncoder
```

```python
def browse_files():
    global model
    model=load(open(r'C:\Users\kandu\OneDrive\Desktop\Mini Project\Code\breachmodel.pkl','rb'))
    global filename
    filename = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("CSV files","*.csv*"),("all files","*.*")))
    label_file_explorer.configure(text="File Opened: "+filename, font = Font_tuple1)

def prediction():
    df=pd.read_csv(filename)
    df.drop(['num_outbound_cmds'], axis=1, inplace=True)
    def le(df):
        for col in df.columns:
            if df[col].dtype == 'object':
                label_encoder = LabelEncoder()
                df[col] = label_encoder.fit_transform(df[col])
    le(df)
    p1=int(e1.get())
    specific_row = df.iloc[p1:p1+1,:]  
    dTDa1 = Toplevel()
    dTDaPT = pt.Table(dTDa1, dataframe=specific_row, showtoolbar=True, showstatusbar=True)
    dTDaPT.show()
    new_data=specific_row
    global breach
    breach=model.predict(new_data)
    if breach==1:
        l3.configure(text='Normal')
    else:
        l3.configure(text="Anomaly")

def close():
    master.destroy()

def update():
    l3.configure(text=' ')

master=Tk()
master.geometry('700x500')
master.title("Breach Prediction")
master.resizable(0,0)

label_file_explorer = Label(master,text = "File Path",width = 100, height = 4,fg = "blue")

button_exit = Button(master,text = "Exit",command = close)
button_search = Button(master,text='Upload file',command=browse_files)
button_predict = Button(master,text='Predict Breach',command=prediction)
button_clear = Button(master,text='Clear',command = update)
e1=Entry(master)
e1.place(x=350,y=205)
l1=Label(master,text="Enter the desired row")
l1.place(x=125,y=200)
label_file_explorer.place(x=-150,y=90)
l3=Label(master,text=" ")
l3.place(x=320,y=350)


Font_tuple = ("Comic Sans MS", 16)
Font_tuple1 = ("Comic Sans MS", 12)
l1.configure(font = Font_tuple)
e1.configure(font = Font_tuple1)
l3.configure(font = Font_tuple)

button_predict.configure(font = Font_tuple)
button_search.configure(font = Font_tuple)
button_exit.configure(font = Font_tuple)
button_clear.configure(font = Font_tuple)

button_search.place(x=300,y=20)
button_exit.place(x=425,y=390)
button_clear.place(x=225,y=390)
button_predict.place(x=275,y=260)

master.mainloop()
```
